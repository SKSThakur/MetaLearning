```python
import pandas as pd 
    
df = pd.read_csv("physionet_A_N.csv") 
   
df.head()
```

```python
#X, Y split for features and labels
X = df.drop(columns="CLASS", axis = 1)
Y = df["CLASS"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#train, test, split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#training for logistic regression
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
#testing using accuracy_score
X_train_prediction = model.predict(X_train)
print("Accuracy on training data = ", accuracy_score(Y_train, X_train_prediction)) 
X_test_prediction = model.predict(X_test)
print("Accuracy on testing data = ", accuracy_score(Y_test, X_test_prediction)) 
```

```python
#observation: low variance is observed for logistic regression
```

```python
#training for a decision tree
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=None)

model.fit(X_train, Y_train)
```

```python
#testing using accuracy_score
X_train_prediction = model.predict(X_train)
print("Accuracy on training data = ", accuracy_score(Y_train, X_train_prediction)) 
X_test_prediction = model.predict(X_test)
print("Accuracy on testing data = ", accuracy_score(Y_test, X_test_prediction)) 
```

```python
#observation: significant overfittting occurs on the above decision tree
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

#using adaboost for training
base_clf = DecisionTreeClassifier(max_depth=1)

model = AdaBoostClassifier(base_estimator=base_clf, n_estimators=50)

model.fit(X_train, Y_train)
```

```python
#testing using accuracy_score
X_train_prediction = model.predict(X_train)
print("Accuracy on training data = ", accuracy_score(Y_train, X_train_prediction)) 
X_test_prediction = model.predict(X_test)
print("Accuracy on testing data = ", accuracy_score(Y_test, X_test_prediction)) 
```

```python
#training random forest
from sklearn.ensemble import RandomForestClassifier

# Create the classifier
model = RandomForestClassifier(n_estimators=100)

# Fit the classifier to the training data
model.fit(X_train, Y_train)
```

```python
#testing using accuracy_score
X_train_prediction = model.predict(X_train)
print("Accuracy on training data = ", accuracy_score(Y_train, X_train_prediction)) 
X_test_prediction = model.predict(X_test)
print("Accuracy on testing data = ", accuracy_score(Y_test, X_test_prediction)) 
```

```python
#ensemble learning techniques are observed to reduce the bias as well as the variance
```
